```python
def normalize(x):
    return [(x[n] - min(x)) / (max(x) - 
           min(x)) for n in range(len(x))]
```

```python
#padronizar variáveis
def scale(X, x_min, x_max):
    nom = (X-X.min(axis=0))*(x_max-x_min)
    denom = X.max(axis=0) - X.min(axis=0)
    denom[denom==0] = 1
    return x_min + nom/denom 
X_scaled = scale(variaveis, -1, 1)
print(X_scaled)


```

```python

inertia = []
for n in range(1 , 20):
    algorithm = (KMeans(n_clusters = n))
    algorithm.fit(X_scaled)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 20) , inertia , 'o')
plt.plot(np.arange(1 , 20) , inertia , '-' , alpha = 0.5)
plt.xlabel('Número de Clusters') , plt.ylabel('Soma das Distâncias Q intra Clusters')
plt.show()
```
